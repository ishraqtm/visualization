Dataset Source: https://open.toronto.ca/dataset/ttc-subway-delay-data/ 

```python
file_path = '/Users/taz/visualization/02_activities/assignments/dsi-ttc-subway-delay-data-2022.xlsx'
df1 = pd.read_excel(file_path, sheet_name='weekend-delay-number')
df2 = pd.read_excel(file_path, sheet_name='weekday-delay-number')
df3 = pd.read_excel(file_path, sheet_name='weekend-delay-duration')
df4 = pd.read_excel(file_path, sheet_name='weekday-delay-duration')

top_10_weekend_delays = df1.nlargest(10, '# DELAYS')
top_10_weekday_delays = df2.nlargest(10, '# DELAYS')


print(df1.head())
print("\n")
print(df2.head())
print("\n")
print(df3.head())
print("\n")
print(df4.head())
```

```python
from IPython.display import Image

print("Figure 1: Weekday and Weekend TTC Subway Delays - Top 10 Stations (2022)")
Image(filename='/Users/taz/visualization/02_activities/assignments/Figure 1 - Top TTC Delays 2022.png', width=800, height=600)
```

```python
plt.figure(figsize=(10, 6))
ax2 = sns.barplot(x='WEEKDAY DELAYS', y='# DELAYS', data=top_10_weekday_delays, palette='RdYlGn_r', hue='AVG DELAY TIME (MINS)')
plt.xlabel('TTC SUBWAY STATION')
plt.ylabel('NUMBER OF DELAYS')
plt.title('Figure 2a - Top 10 Stations with Most Weekday Delays (2022)')
get_labels = plt.xticks(rotation=45, ha='right')
plt.tight_layout()

for container in ax2.containers:
    ax2.bar_label(container)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='WEEKEND DELAYS', y='# DELAYS', data=top_10_weekend_delays, palette='RdYlGn_r', hue='AVG DELAY TIME (MINS)')
plt.xlabel('TTC SUBWAY STATION')
plt.ylabel('NUMBER OF DELAYS')
plt.title('Figure 2b - Top 10 Stations with Most Weekend Delays (2022)')
get_labels = plt.xticks(rotation=45, ha='right')
plt.tight_layout()

for container in ax.containers:
    ax.bar_label(container)
```
