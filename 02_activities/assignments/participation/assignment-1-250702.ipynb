```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set seed for reproducibility
np.random.seed(613)

# Generate data
x = np.arange(50)
y = np.random.randint(0, 100, 50)

# Create figure and axis
fig, ax = plt.subplots(figsize=(5, 3))
#ax.scatter(x,y)
#ax.bar (x,y)
#ax.barh(x,y)
#ax.plot(x,y)
ax.hist(y)
```

```python
# Define a font style dictionary
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# Create fig and ax with size 5 inches by 3 inches 
fig, ax = plt.subplots(figsize=(5, 3))
# Plot x and y data 
ax.plot(x,y)

# Set custom title and labels with font styles
ax.set_title('Total growth over time', fontdict = font1,loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
#added markers and colors
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y,
           marker = '*',
           color = 'indigo')
```

```python
#Add line width and linestyle
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
           y,
           marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2)
```

```python
#Add custom markers and colors
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2,
           markersize=12,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')
```

```python
#Add grid lines with custom styling
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2,
           markersize=12,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')
ax.grid(axis='y', color = 'blue', linewidth =2, linestyle='-.')
```
