```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
#Make a line plot with both y on same axes
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
#Add legend

fig, ax = plt.subplots(figsize = (5,3))
ax.plot (x, y1, label = "Person 1")
ax.plot (x, y2, label = "Person 2")
ax.legend (loc = "lower right")
```

```python
#Plot both y variables on the same axis
fig, ax = plt.subplots(figsize = (5,3))
ax.plot (x, y1, label = "Person 1")
ax.plot (x, y2, label = "Person 2")
ax.legend (loc = "lower right",
           frameon = True, #add frame around the legend
           fontsize = 12,
           ncol = 2, #specify number of columns
           shadow = True)
```

```python
# Move the legend outside of the plot area
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label='Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc = 'upper left',
          bbox_to_anchor=(1, 1))
```

```python
# Annotations added
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.text(10, 95, "This value is important!")
```

```python
#Modify text annotations
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.text(10, 95, "This value is important!",
        ha = 'center',
        color = 'red',
        size = 20) 
```

```python
#Position text annotations
fig, ax = plt.subplots()

ax.axis([0,10,0,10])

ax.text(1, 5, ".Data:(1,5)", transform = ax.transData)
ax.text(0.5, 0.1, ".Axes:(0.5,0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, ".Figure: (0.2, 0.2)", transform = fig.transFigure)
```

```python
#Annotate with arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
    arrowprops = dict(arrowstyle = "wedge",
                      color = "hotpink"))
```

```python
#Removing tick marks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
#Limiting tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')
plt.xticks(rotation=45, ha='right')
```

```python
# Modify axis title fonts
font1 = {'family': 'serif', 'color': 'indigo'}
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
plt.xlabel("Shiny New X Axis!", fontsize = 18, fontdict = font1)
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7, 3))

ax1.scatter(x1, y1)
ax2.bar(x2, y2)
```

```python
font1 = {'family': 'serif', 'size': 10}
font2 = {'family': 'serif', 'size': 14}

fig, (ax1, ax2) = plt.subplots(ncols = 2,  
                               nrows = 1,
                               figsize = (9,3))
ax1.scatter(x1,y1,
            marker = 'o',
            color = 'red')
ax2.bar(x2,y2, color = 'darkgreen')

ax1.set_title('Scatterplot', fontdict = font2)
ax1.set_ylabel('Measure', fontdict = font1)
ax1.set_xlabel('Time', fontdict = font1)
ax2.set_title('Pirate Strength', fontdict = font2)
ax2.set_ylabel('Strength', fontdict = font1)
ax2.set_xlabel('Names', fontdict = font1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],
                                   figsize = (7,4),
                                   layout= "constrained")

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```
